```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("C:\\Users\\welcome\\Downloads\\API_SP.POP.TOTL_DS2_en_csv_v2_127006.csv", skiprows=4)

print(df.head())
print(df.columns)

india_data = df[df['Country Name'] == 'India']

years = india_data.columns[4:]
population_values = india_data[years].values.flatten()

plt.figure(figsize=(12,6))
plt.bar(years, population_values, color='skyblue')
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("India's Population over Years")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

np.random.seed(42)
ages = np.random.randint(18, 80, size=200)

plt.figure(figsize=(8,5))
plt.hist(ages, bins=10, color='lightgreen', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Number of people")
plt.title("Distribution of Ages in a Sample Population")
plt.tight_layout()
plt.show()

genders = np.random.choice(['Male', 'Female', 'Other'], size=200, p=[0.48, 0.48, 0.04])

gender_counts = pd.Series(genders).value_counts()

plt.figure(figsize=(6,4))
gender_counts.plot(kind='bar', color='coral')
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Genders in a Sample Population")
plt.tight_layout()
plt.show()
```
